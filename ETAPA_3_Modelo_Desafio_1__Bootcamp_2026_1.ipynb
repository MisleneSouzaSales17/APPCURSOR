DESAFIO 1

Você está ajudando a organizar uma conferência de mulheres na tecnologia e precisa criar crachás com cores diferentes para facilitar o networking. Crie um programa que solicita à participante quantos anos de experiência ela possui e, em seguida, atribui uma categoria:

* **0 a 2 anos:** Crachá Verde (Júnior)
* **3 a 6 anos:** Crachá Amarelo (Pleno)
* **7 anos ou mais:** Crachá Vermelho (Sênior)

```python
# Programa para atribuir categoria de crachá em uma conferência

# Solicita os anos de experiência da participante
anos_experiencia = float(input("Quantos anos de experiência você possui? "))

# Verifica a categoria de acordo com os anos informados

if anos_experiencia >= 0 and anos_experiencia <= 2:
  print("Cracha Verde - Junior")
elif anos_experiencia >= 3 and anos_experiencia <= 6:
  print("Cracha Amarelo - Pleno")
elif anos_experiencia >= 7:
  print("Cracha Vermelho - Senior")
else:
  print("Valor Inválido. Digite um número inteiro.")
```

DESAFIO 2

Durante o bootcamp, as participantes registram o tempo total de estudo em minutos. Para facilitar a análise, você precisa converter esse valor para horas e minutos.

Crie um programa que leia um número inteiro representando o total de minutos estudados e exiba o valor equivalente em horas e minutos.

```python
# Programa para converter minutos de estudo em horas e minutos
# Solicita o total de minutos estudados

total_minutos = int(input("Digite o total de minutos estudados: "))
# Calculo horas e minutos

horas = total_minutos // 60
minutos = total_minutos % 60

# Resultado
print(f"{total_minutos} minutos equivalem a {horas} horas e {minutos} minutos.")
```

Pergunta 1: "Você gosta de criar relatórios e dashboards?"

Pergunta 2: "Você se interessa por estatística e modelos preditivos?"

Pergunta 3: "Você gosta de trabalhar com a infraestrutura e o fluxo de dados?"

Se a resposta para a pergunta 1 for "sim", sugira "Trilha de Analista de BI".

Se a resposta para a pergunta 3 for "sim", sugira "Trilha de Engenheira de Dados".

```python
# Programa de sugestão de trilha de mentoria
# Faz as perguntas para a usuária

pergunta1 = input("Você gosta de criar relatórios e dashboards? (sim/não): ")
pergunta2 = input("Você se interessa por estatística e modelos preditivos? (sim/não): ")
pergunta3 = input("Você gosta de trabalhar com a infraestrutura e o fluxo de dados? (sim/não): ")

# Verifica as respostas e sugere a trilha

if pergunta1 == "sim":
  print("Trilha de Analista de BI")
elif pergunta2 == "sim":
  print("Trilha de Cientista de Dados")
elif pergunta3 == "sim":
  print("Trilha de Engenheira de Dados")
else:
   print("Trilha de Fundamentos de Dados")
```

DESAFIO 4

Crie um pequeno sistema para gerenciar uma lista de artigos sobre tecnologia escritos por mulheres. O sistema deve ser um menu que roda em um loop `while` e oferece as seguintes opções:

1. **Adicionar artigo:** Pede à usuária o título, a autora e uma tag (ex: 'Python', 'Carreira').
2. **Listar todos os artigos:** Exibe todos os artigos cadastrados.
3. **Buscar por autora:** Pede o nome de uma autora e exibe todos os artigos escritos por ela.
4. **Sair**

Use uma lista de dicionários para armazenar os artigos.

```python
# Sistema de gerenciamento de artigos sobre tecnologia escritos por mulheres

# Lista para armazenar os artigos
artigos = []

while True:
    print("\n--- MENU ---")
    print("1. Adicionar artigo")
    print("2. Listar todos os artigos")
    print("3. Buscar por autora")
    print("4. Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        titulo = input("Digite o título do artigo: ")
        autora = input("Digite o nome da autora: ")
        tag = input("Digite a tag do artigo (ex: Python, Carreira): ")

        artigo = {"titulo": titulo, "autora": autora, "tag": tag}
        artigos.append(artigo)
        print("✅ Artigo adicionado com sucesso!")

    elif opcao == "2":
        if len(artigos) == 0:
            print("Nenhum artigo cadastrado ainda.")
        else:
            print("\n--- Lista de Artigos ---")
            for i, artigo in enumerate(artigos, start=1):
                print(f"{i}. {artigo['titulo']} - {artigo['autora']} (Tag: {artigo['tag']})")

    elif opcao == "3":
        nome_autora = input("Digite o nome da autora: ")
        encontrados = [a for a in artigos if a["autora"].lower() == nome_autora.lower()]

        if len(encontrados) == 0:
            print("Nenhum artigo encontrado para essa autora.")
        else:
            print(f"\n--- Artigos de {nome_autora} ---")
            for artigo in encontrados:
                print(f"- {artigo['titulo']} (Tag: {artigo['tag']})")

    elif opcao == "4":
        print("Encerrando o sistema. Até logo!")
        break

    else:
        print("Opção inválida. Tente novamente.")

```

Durante o processo de limpeza de dados, você precisa testar a manipulação de strings.

Crie um programa que receba uma palavra e exiba essa palavra invertida
(exemplo: "python" → "nohtyp").

```python
# Programa para inverter uma palavra

# Solicita a palavra
palavra = input("Digite uma palavra: ")

# Inverte a palavra usando slicing
palavra_invertida = palavra[::-1]

# Exibe o resultado
print(f"A palavra invertida é: {palavra_invertida}")
```
