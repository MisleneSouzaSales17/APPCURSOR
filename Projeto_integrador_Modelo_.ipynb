```python
#Co-Work Segunda Casa - Ambiente aconchegante e inovador
#Co-Work Tech-Nature - Seu espaço de trabalho em meio a natureza
class ReservadeSala:
    def __init__(self):
        self.reservas = []

    def registrar_reserva(self, sala, horario):
        if not self.consultar_disponibilidade(sala, horario):
            self.reservas.append({'sala': sala, 'horario': horario})
            print("Reserva registrada.")
        else:
            print("Horário já reservado.")

    def cancelar_reserva(self, sala, horario):
        for reserva in self.reservas:
            if reserva['sala'] == sala and reserva['horario'] == horario:
                self.reservas.remove(reserva)
                print("Reserva cancelada.")
                return
        print("Reserva não encontrada.")

    def visualizar_reservas(self):
        if not self.reservas:
            print("Nenhuma reserva registrada.")
            return
        for reserva in self.reservas:
            print(f"Sala: {reserva['sala']}, Horário: {reserva['horario']}")

    def consultar_disponibilidade(self, sala, horario):
        for reserva in self.reservas:
            if reserva['sala'] == sala and reserva['horario'] == horario:
                return True
        return False

def menu():
    sistema = ReservaSala()
    while True:
        print("\n1. Registrar Reserva")
        print("2. Cancelar Reserva")
        print("3. Visualizar Reservas")
        print("4. Consultar Disponibilidade")
        print("5. Sair do Sistema")
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            sala = input("Sala: ")
            horario = input("Horário: ")
            sistema.registrar_reserva(sala, horario)
        elif opcao == 2:
            sala = input("Sala: ")
            horario = input("Horário: ")
            sistema.cancelar_reserva(sala, horario)
        elif opcao == 3:
            sistema.visualizar_reservas()
        elif opcao == 4:
            sala = input("Sala: ")
            horario = input("Horário: ")
            if sistema.consultar_disponibilidade(sala, horario):
                print("Horário indisponível.")
            else:
                print("Horário disponível.")
        elif opcao == 5:
            break
        else:
            print("Opção inválida.")

if __name__ == "__main__":
    menu()
```
